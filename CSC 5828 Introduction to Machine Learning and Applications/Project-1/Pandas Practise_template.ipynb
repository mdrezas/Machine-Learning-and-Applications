##### 3. Create  a DataFrame d based on the given dict. 
#### (1) Print the shape of d.
#### (2) Print the type of values in d.
#### (3) Print the index of d.
#### (4) Print the columns of d.
#### (5) Print the summary of d (use function descibe()).

```python
# Create a DataFrame d based on the given dict.
dic = {
        'name':['Andy','James','Lucy'],
        'age':[18,20,22],
        'gender':['male','male','female'] 
       }
```

```python
d=pd.DataFrame(dic)
```

```python
# Print the shape of d.
print(d.shape)
```

```python
# Print the type of values in d.
print(d.dtypes)
```

```python
# Print the index of d.
for row in d.index:
    print(row, end = " ")
```

```python
#'Print the columns of d.
list(d.columns)
```

```python
# Print the summary of d (use function descibe()).
print(d.describe())
```
