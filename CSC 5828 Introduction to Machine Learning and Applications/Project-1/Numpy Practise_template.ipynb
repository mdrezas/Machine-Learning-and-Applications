#### 2. Create  a one-dimensional array a and initialize as [4,5,6]. 
#### (1) Print the type of a
#### (2) Print the shape of a
#### (3) Print the first element in a (the value should be 4)

```python
print('Create a one-dimensional array a and initialize as [4,5,6]')
a = np.array([4, 5, 6])
print('\n')

print('Print the type of a')
print(type(a))
print('\n')

print('Print the shape of a')
print(np.shape(a))
print('\n')

print('Print the first element in a (the value should be 4)')
print(a[0])
```

#### 4.

```python
print('Create a matrix a, which is all 0, of size 3x3')
dimensions = (3, 3)
a=np.zeros(dimensions) 
print(a)
print('\n')

print('Create a matrix b, which is all 1, of size 4x5')
dimensions = (4, 5)
b=np.ones(dimensions) 
print(b)
print('\n')

print('Create a unit matrix c, of size 4x4')
c=np.identity(4)
print(c)

print('\n')

print('Create a random matrix d, of size 3x2')
dimensions = (3, 2)
d=np.random.rand(3,2)
print(d)

```

#### 7. 
#### (1) Create  an array x and initialize as [[1, 2], [3, 4]], dtype=np.float64.
#### (2) Create  an array y and initialize as [[5, 6], [7, 8]], dtype=np.float64.
#### (3) Print x + y and np.add(x, y)
#### (4) Print x - y and np.subtract(x, y）
#### (5) Print x * y, np.multiply(x, y）and np.dot(x,y), and compare the results
#### (6) Print x / y and np.divide(x,y)
#### (7) Print  the sqaure of x (hint: np.sqrt())
#### (8) Print x.dot(y) and np.dot(x,y)

```python
print('(1) Create an array x and initialize as [[1, 2], [3, 4]], dtype=np.float64.')
x = np.array([[1,2], [3,4]], dtype=np.float64)
print(x)
print('\n')

print('(2) Create an array y and initialize as [[5, 6], [7, 8]], dtype=np.float64.')
y = np.array([[5,6], [7,8]], dtype=np.float64)
print(y)
print('\n')

print('(3) Print x + y and np.add(x, y)')
print(x+y)
print(np.add(x,y))
print('\n')

print('(4) Print x - y and np.subtract(x, y）')
print(x-y)
print(np.subtract(x,y))
print('\n')

print('(5) Print x * y, np.multiply(x, y）and np.dot(x,y), and compare the results')
print(x*y)
print(np.multiply(x,y))
print(np.dot(x,y))
print('\n')

print('x*y/np.multiply does an element-wise multiplication, while np.dot is the dot product of two matrices. Therefore, there is differences in out too.')
print('(6) Print x / y and np.divide(x,y)')
print(x/y)
print(np.divide(x,y))
print('\n')

print('(7) Print the sqaure of x (hint: np.sqrt())')
print(np.square(x))
print('\n')

print('(8) Print x.dot(y) and np.dot(x,y)')
print(x.dot(y))
print(np.dot(x,y))
```

#### 12. Plot sin() and cos()   (hint: use np.sin(), np.cos() and matplotlib.pyplot package）
#### (1) axis X: x = np.arange(0, 3 * np.pi, 0.1)   axis Y: y=sin(x)
#### (2) axis X: x = np.arange(0, 3 * np.pi, 0.1)   axis Y: y=cos(x)

```python
print('Plot SIN wave')
import matplotlib.pyplot as plt
x = np.arange(0, 3 * np.pi, 0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Sin Wave")
plt.show()
```

```python
print('Plot COS wave')
x = np.arange(0, 3 * np.pi, 0.1)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Cos Wave")
plt.show()
```
